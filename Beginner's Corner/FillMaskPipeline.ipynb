```python
from transformers import pipeline
import pandas as pd
import numpy as np
import textwrap
```

```python
df = pd.read_csv('bbc_text_cls.csv')
```

```python
df.head()
```

```python
set(df['labels'])
```

```python
label = 'business'
```

```python
texts = df[df['labels'] == label].text
```

```python
np.random.seed(1234)
```

```python
i = np.random.choice(texts.shape[0])
doc = texts.iloc[i]
```

```python
print(textwrap.fill(doc, replace_whitespace=False, fix_sentence_endings=False))
```

```python
mlm = pipeline('fill-mask')
```

```python
mlm('Quiksilver <mask> for Rossignol')
```

```python
mlm("<mask> of Skis Rossignol, the world's \
largest ski-maker, have jumped as much as 15% on speculation that it \
will be bought by US surfwear firm Quiksilver.")
```

```python
mlm("Shares of Skis Rossignol, the world's \
largest ski-maker, have jumped as much as 15% on <mask> that it \
will be bought by US surfwear firm Quiksilver.")
```
